## Detecting Outliers

```python
def remove_outliers_iqr(df, column_name):
    # Calculate the first quartile (Q1) and third quartile (Q3) of the specified column
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)

    # Compute the Interquartile Range (IQR)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Return a DataFrame with values within the bounds (i.e., without outliers)
    return df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
```

### It filters out the outliers from the dataframe based on the IQR , it removes values that are more than 1.5times the IQR below the 1st quartile or above the 3rd quartile.

```python
outlier_cols = ['Account_Length','Day_Mins','Day_Calls', 'Eve_Mins', 'Eve_Calls', 'Night_Mins','Night_Calls']
# Loop through each selected col/feature
for col in outlier_cols:
    df = remove_outliers_iqr(df, col)

# Resetting index after filtering else it'll be inconsistent!
df.reset_index(drop=True, inplace=True)
```

### The code removes outliers from specified columns in the DataFrame and then resets the index to ensure it is sequential and consistent.

## Import necessary libraries for data preprocessing, clustering, and visualization

```python
from sklearn.preprocessing import StandardScaler  # For standardizing features
from sklearn.cluster import KMeans  # For applying K-means clustering
from sklearn.decomposition import PCA  # For dimensionality reduction using Principal Component Analysis
import matplotlib.pyplot as plt  # For plotting and visualizing data
```

```python
# Ensure 'Churn' column is properly formatted
df['Churn'] = df['Churn'].astype(int)
```

### This mainly converts the datatype of churn column into integer(i.e if the value of churn is 1 it means that the customer who has churned-left the service , if it is 0 then the customer who has not churned- retained )

```python
df.columns 
```

```python
numerical_cols = [
    'Account_Length', 'VMail_Message', 'Day_Mins', 'Day_Calls', 'Day_Charge', 
                    'Eve_Mins', 'Eve_Calls', 'Eve_Charge', 'Night_Mins', 'Night_Calls', 
                    'Night_Charge', 'Intl_Mins', 'Intl_Calls', 'Intl_Charge', 'CustServ_Calls'
]
X = df[numerical_cols]
# Create a df that contains the list of numeric columns.
```

### Initialize a StandardScaler object and apply it to the feature matrix X to standardize the features 
### (mean = 0, variance = 1) by removing the mean and scaling to unit variance.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Calculate the Within-Cluster Sum of Squares (WCSS) for different numbers of clusters (1 to 10)
### using the K-Means algorithm. This will help in determining the optimal number of clusters 
### by analyzing the "elbow" in the WCSS plot.

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

### Helps to identify the elbow point to optimise the clusters from the given range.

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Identified the elbow point occurs at 3.So lets assign the optimal cluster as 3.

```python
# Choose the optimal number of clusters (e.g., 3)
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# Visualize the clusters using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]
```

```python
# Calculate Customer Lifetime Value (CLV)
df['CLV'] = df['Day_Charge'] + df['Eve_Charge'] + df['Night_Charge'] + df['Intl_Charge']
```

```python
# Analyze average CLV for each cluster
cluster_clv = df.groupby('Cluster')['CLV'].mean()
print("\nAverage CLV for each cluster:")
print(cluster_clv)
```

```python
# Assigning cluster colors directly based on expected cluster numbers
df['Cluster_Color'] = df['Cluster'].map({
    0: 'Potential High Value Customers', 
    1: 'Potential Low Value Customers', 
    2: 'Potential Medium Value Customers'
})
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='PCA1', y='PCA2', hue='Cluster_Color', data=df, 
    palette=['#32CD32', '#4682B4','#FF6347' ], s=100
)
plt.title('Cluster Visualization using PCA')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend(title='Upselling Potential', loc='best')
plt.show()
```

```python
# Analyze clusters for upselling opportunities
print("\nCluster Analysis:")
for i in range(optimal_clusters):
    print(f"\nCluster {i} statistics:")
    print(df[df['Cluster'] == i][numerical_cols].describe())
```

```python
# Calculate and print CLV for each cluster
print("\nCluster-wise CLV Analysis:")
for i in range(optimal_clusters):
    cluster_clv_value = df[df['Cluster'] == i]['CLV'].mean()
    print(f"Average CLV for Cluster {i}: {cluster_clv_value:.2f}")
# Count the number of records in each cluster
cluster_counts = df['Cluster'].value_counts()
```

```python
# Print the number of records in each cluster
print("\nNumber of records in each cluster:")
print(cluster_counts)
```
