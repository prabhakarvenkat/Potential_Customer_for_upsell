#### Dataset Description

The dataset appears to be a Call Detail Record (CDR) containing various details about phone calls made by customers. The columns in the dataset are as follows:

1. *Phone Number*: Unique identifier for each customer.
2. *Account Length*: Duration (in days) for which the account has been active.
3. *VMail Message*: Number of voice mail messages.
4. *Day Mins*: Total minutes of calls made during the day.
5. *Day Calls*: Total number of calls made during the day.
6. *Day Charge*: Total charge for calls made during the day.
7. *Eve Mins*: Total minutes of calls made during the evening.
8. *Eve Calls*: Total number of calls made during the evening.
9. *Eve Charge*: Total charge for calls made during the evening.
10. *Night Mins*: Total minutes of calls made during the night.
11. *Night Calls*: Total number of calls made during the night.
12. *Night Charge*: Total charge for calls made during the night.
13. *Intl Mins*: Total minutes of international calls.
14. *Intl Calls*: Total number of international calls.
15. *Intl Charge*: Total charge for international calls.
16. *CustServ Calls*: Number of calls made to customer service.
17. *Churn*: Indicator of whether the customer has churned (True/False).
